#10.종량제 붕투 가격

* 20L 가격과 5L 가격이 모두 0원이 아닌 데이터만 필터를 한 후, 각row별로 20L가격과 5L 가격의 차이를 '차이가격'이라 부른다고 하지. 시도명 별 차이가격의 평균가격을 비교할때 그 값이 가장 큰 금액을 반올림하여 정수로 답하여

```python
#1. 데이터 불러오기
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/refs/heads/main/waste_bags.csv')
df.head(3)
```

```python
# 2. 조건 필터링: '20L 가격'과 '5L 가격'이 모두 0이 아닌 행만 필터링
temp = df[(df['20L가격'] != 0) & (df['5L가격'] != 0)]

# 3. 가격 차이 계산: '20L 가격' - '5L 가격'
temp['차이가격'] = temp['20L가격'] - temp['5L가격']

# 4. 그룹화 및 평균 계산: '시도별'로 그룹화하고 '가격차이'의 평균 계산
B = temp.groupby('시도명')['차이가격'].mean().sort_values(ascending=False)

# 5. 가장 큰 평균 차이를 가지는 값 찾기
print(round(B.max()))
```

#11.성인 체중 및 키 데이터


* BMI는 몸무게(kg) / (키(M) * 키(M))로 정의 된다. 초고도 비만은 BMI 25이상, 고도비만은 BMI 25미만-23이상, 정상은 23미만 - 18.5이상 저제중은 18.5미만으로 정의된다. 주어진 데이터에서 초고도비만 인원+저체증 인원의 숫자는? *이탤릭체 텍스트*

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/refs/heads/main/height_weight.csv')
df.head(3)
```

```python
# 2. 키를 미터 단위로 변환
df['height(m)'] = df['height(cm)']/100

# 3. BMI 계산: BMI = 몸무게(kg) / (키(m) * 키(m))
df['BMI'] = df['weight(kg)']/(df['height(m)']*df['height(m)'])

# 4. 초고도비만과 저체중 조건 정의
High_BMI = df['BMI'] >= 25
Low_BMI = df['BMI'] < 18.5

# 5. 초고도비만과 저체중에 해당하는 사람의 수 계산
result = df[High_BMI | Low_BMI]
print(len(result))
```

#12.년도별 서울 각 구의 초,중,고 전출 전입 인원

* 순유입인원은 초중고 도내,도외 전입인원에서 초중고 도내,도외 전출인원을 뺀값이다. 각년도별로 가장 큰 순유입인원을 가진 지역구의 순유입인원을 구하고 전체 기간의 해당 순유입인원들의 합을 구하여라

```python
import pandas as pd

# 1. CSV 파일 읽기
df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/refs/heads/main/e5_p1_3.csv')
df.head()
```

```python
# 2. 초등, 중등, 고등 전입/전출 합계 계산
# 초등, 중등, 고등 전입 합계
A = df.loc[:,df.columns.str.contains('전입')].sum(axis=1)
# 초등, 중등, 고등 전출 합계
B = df.loc[:,df.columns.str.contains('전출')].sum(axis=1)

```

```python
data = {
    '지역': df['지역'],
    '전입': A,
    '전출': B,
    '년도': df['년도']
}

a = pd.DataFrame(data)

# 3. 순유입인원 계산 (전입 - 전출)
a['순유입인원'] = a['전입'] - a['전출']

# 4. 년도별로 가장 큰 순유입 인원을 가진 지역의 순유입인원 합 계산
result = a.groupby('년도')['순유입인원'].max().sum()

print(result)
```

#14. 초등학교 학생,교사 수 데이터

```python
# 1. 데이터 읽기
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/refs/heads/main/big_6_1_2.csv')
df.head(3)
```

#15.범죄 발생 수 데이터

* 연도별로 총 범죄건수( 범죄유형별 발생 건수의 합)의 월평균 값을 구한 후, 그 값이 가장 큰 연도를 찾아, 해당 연도의 총 범죄 건수의 월평균 값을 반올림하여 정수로 출력하세요

```python
# 1. 데이터 읽기
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/refs/heads/main/big_6_1_3.csv')
df.head(3)
```

```python
# 2. '년도' 열 생성 (datetime 변환 후 연도 추출)
df['년도'] = df['월별'].astype('datetime64[ns]').dt.year

# 3. 범죄 유형별 발생 건수 합산하여 '총합계' 열 생성
df['총범죄']= df['강력범죄'] + df['교통범죄'] + df['사이버범죄'] + df['지능범죄'] + df['마약범죄'] + df['경범죄']

# 4. 월 별 총 범죄 건수 합계 계산
month = df.groupby('년도')['총범죄'].sum()/12

# 5. 가장 큰 연도의 총 범죄 건수를 월평균으로 계산
result = round(month.sort_values(ascending=False).iloc[0])
print(result)
```

#16.결측치,표준화,최댓값


* student_scores.csv를 사용하여 score에서 발생한 결측치를 제거하고, 학생들이 가장 많은 학생들이 들었던 수업 과목에서 해당 과목의 점수를 표준화하여 가장 놓은 점수를 구하여 출력합니다.(반올림하여 소수점아래 3자리까지 출력)

```python
# 1. 데이터 읽기
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/refs/heads/main/student_scores.csv')
df.head(3)
```

```python
from sklearn.preprocessing import StandardScaler

# 2. 결측치 제거
df = df.dropna(subset=['score'])

#3. 학생들이 가장 많이 선택한 과목 찾기
max = df['subject'].value_counts().index[0]

# 4. 해당 과목의 점수 데이터만 선택
scores = df.loc[df['subject'] == max,['score']].copy()

# 5. 점수 표준화
result = StandardScaler().fit_transform(scores).max()

# 6. 결과 출력
print(round(result,3))
```

#18.이상치 개수

* outlier_data.csv의 feature_2칼럼내의 이상치 개수를 구하여라
* 이상치<평균 - IQR+1.5, 이상치>평균 + IQR*1.5


```python
# 1. 데이터 읽기
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/refs/heads/main/outlier_data.csv')
df.head(3)
```

```python
# 1. 데이터프레임에서 'feature_2' 열 선택
data=df['feature_2']

# 2. Q1과 Q3 계산
Q1, Q3 = data.quantile([0.25,0.75])

# 3. IQR 계산
IQR = Q3 - Q1

# 4. 이상치 기준 계산
lower = data.mean() - IQR * 1.5
upper = data.mean() + IQR * 1.5

# 5. 이상치 조건 정의
outlier = (data<lower) | (data>upper)

# 6. 이상치 개수 출력
print(outlier.sum())
```
