### 4-05.연습문제2 - 학점구하기

````python
# 0~100점의 점수에 대해 다음 기준을 적용하여 학점을 출력한다
# 100 ~ 90 : A
# 89 ~ 80 : B
# 79 ~ 70 : C
# 69 ~ 60 : D
# 59 ~ 0 : F
# 단, 0점 미만, 100점 초과 점수에 대해서는 “Error” 를 출력한다
````

````python
# if ~ elif ~ elif ~ else 사용
score = 89
if 90 <= score <=100:
  result ="A"
elif 80 <= score <90:
  result ="B"
elif 70 <= score <80:
  result ="C"
elif 60 <= score <70:
  result ="D"
elif 0 <= score <60:
  result ="F"
else:
  result = "Error"


print(result)
````

````python
# str, subscription 사용
score = 99
grade = "FFFFFFDCBAA"    #   78 // 10 -> 7, 92 // 0 -> 9
if score < 0 or score > 100:
    result = "Error"
else:
    result = grade[score//10]
print(result)
````

````python
# str, subscription 사용
score = 99
grade = "FFFFFFDCBAA"    #   78 // 10 -> 7, 92 // 0 -> 9
result = "Error" if score < 0 or score > 100 else grade[score//10]
print(result)
````

### 4-06.for의 사용

            statements

- iterable : list, tuple, str, set, dict ... + iterator, generator        

````python
for target_list in iterable:
    statement (target_list를 활용)
# 1. in 뒤에 있는 대상에 대해 반복할 수 있는 형태로 변환한다
# 2. 반복대상에서 다음 item을 가져와서 이름(target_list)을 붙인다
# 3. 이름 붙은 item을 사용해 statement를 실행
# 4. 다시 2번을 수행, 이 작업을 하다가 더이상 데이터가 없으면 반복을 종료
````

````python
# [1] for문을 사용하여 colors의 item을 한 행에 하나씩 출력한다
colors = ['red', 'green', 'blue']
for color in colors:
  print(color)
````

### 4-08.for문 연습문제 1(합격, 불합격)
- 총 5명 학생의 성적이 다음과 같을 때 시험 점수가 60점 이상이면 "합격입니다."를 그렇지 않으면 "불합격입니다."를 점수와 함께 출력하세요.
- 사용) result = [90, 35, 68, 44, 81]

````python
result = [90, 35, 68, 44, 81]
for x in result:
  if x >= 60:
    msg = '합격입니다'
  else:
    msg = '불합격입니다'
  print(f'{x} - {msg}')

````

### 4-10.for 고급

````python
# [1] for문을 사용하여 다음 a의 내용을 아래와 같이 출력하라
# 1 2
# 3 4
# 5 6
a = [(1, 2), (3, 4), (5, 6)]
for x,y in a:
  print(x,y)
````

````python
a = [(1, 2), (3, 4), (5, 6)]
for x in a:
  print(*x)
````

````python
# [2] for문과 zip을 사용하여 아래와 같이 출력하라
# A 100
# B 200
# C 300
name = ['A', 'B', 'C']
unit = [100, 200, 300]
for x,y in zip(name,unit):
  print(x,y)
````

````python
# [3] for문과 enumerate을 사용하여 data의 값을 * 2 한 값으로 변경하여 출력하라
#     [20, 40, 60] 으로 수정하여 출력하도록 한다.
data = [10, 20, 30]
for a,b in enumerate(data):
  data[a] = b*2
print(data)
````

````python
data = [10, 20, 30]
for i in range(len(data)):    # 0 1 2
    data[i] *= 2
print(data)
````

### 4-11. break, continue
- break : 반복을 종료
- continue : 반복할 나머지 문장을 수행하지 않고 계속 반복 수행
  (다음 item을 가져오는 동작으로 바로 이동)

````python
# [1] a에 있는 데이터를 출력하다가 데이터가 0보다 작은 경우 종료되도록 한다 (break 사용)
a = [1, 5, -1, 2, -3, 4]
for x in a:
  if x<0:
    break
  print(x)
````

````python
# [2] a에 있는 데이터에서 양수만 출력되도록 한다 (continue 사용)
b = [1, 5, -1, 2, -3, 4]
for x in b:
  if x<0:
    continue
  print(x)
````

````python
a = [1, 5, -1, 2, -3, 4]
for x in a:
    if x >= 0:
        print(x)
````
