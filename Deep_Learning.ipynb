#1-2 머신러닝 소개

• 지도 학습 (supervised learning)

• 입력(input, X, 독립변수, 설명변수)과 타깃(target(output), Y, 종속변수, 반응변수)이 모두 있음

• 학습 데이터 (training data) : 모형 구축을 위하여 사용되는 데이터

• 테스트 데이터 (test data) : 구축된 모형의 평가를 위하여 사용되는 데이터

• 비지도 학습 (unsupervised learning) = 자율학습

• 타깃(target)이 없는 데이터에 대한 학습 → 군집화 (clustering)

Supervised learning와 unsupervised learning 비교


*   고객의 신용 상태 예측
*   지도 학습 : 연봉과 신용상태 이용 가능
  *   P(신용상태=Good) = 2.48 + 3.17*연봉
*   비지도 학습 : 연봉 데이터만 이용 가능
  *   연봉을 이용한 군집화

강화 학습(reinforcement learning)
  *   머신러닝 알고리즘을 이용하여 에이전트(agent)를 훈련(training)
  *   훈련된 에이전트는 주어진 환경(enviornment)에서 자신의 판단방식(policy)에 따라 행동을 수행하고 수행결과에 따라 보상(reward)을 받을 수 있음
  *   에이전트의 목표는 최대한 많은 보상을 받는 것
  *   강화학습의 예
    *   Q-Learning,Deep Q leearning

가중치(기울기)와 절편을 구하면 하나의 규칙이 생성 됨

Model = 모델,모형
  *   Training data로부터 구축된 머신러닝 결과(규칙)

모수 (model parameter)
  *   가중치(기울기),절편

손실함수 (loss function)
  *   실제값과 에측값의 차이를 계산하는 함수
  *   손실함수를 이용하여 모수를 수정(update)
  *   손실함수가 최소가 될 때 최적 모형이 됨(미분 이용)

```python
# List 복습
my_list = [10,'hello list',20]
print(my_list[1])

my_list2 = [[10,20,30],[40,50,60]]
print(my_list2[1][1]) # 파이썬의 인덱스는 0부터 시작

import numpy as np
print(np.__version__) #넘파이 버전 확인

my_arr = np.array([[10,20,30],[40,50,60]])
print(my_arr) # 2차원 배열 만들기

type(my_arr) # 넘파이 배열인지 확인

my_arr[0][2] # 특정 배열 요소 선택

np.sum(my_arr) # 넘파이 내장함수 사용
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randn(1000) # 표준 정규 분포를 따르는 난수 1,000개를 만듭니다.
y = np.random.randn(1000) # 표준 정규 분포를 따르는 난수 1,000개를 만듭니다.
plt.scatter(x, y)
plt.show() # 넘파이 배열로 산점도 그리기
```

```python
# X~Poisson(5), Y=2*X+Normal(3,4)
import matplotlib.pyplot as plt
import numpy as np
X = np.random.poisson(5,100)
Y = 2*X+np.random.normal(3,2,100) # 평균=3, 표준편차=2
print(X)
plt.scatter(X, Y)
plt.show() # 넘파이 배열로 산점도 그리기
```
