Ïò§Ï¶àÎπÑ (OR)
*   pÍ∞Ä 0Î∂ÄÌÑ∞ 1ÍπåÏßÄ Ï¶ùÍ∞ÄÌï† Îïå Ïò§Ï¶àÎπÑÏùò Í∞íÏùÄ Ï≤òÏùåÏóêÎäî Ï≤úÏ≤úÌûà Ï¶ùÍ∞ÄÌïòÏßÄÎßå pÍ∞Ä 1Ïóê Í∞ÄÍπåÏõåÏßÄÎ©¥ Í∏âÍ≤©Ìûà Ï¶ùÍ∞Ä



```python
# Ïò§Ï¶àÎπÑ Ïã§Ïäµ
import matplotlib.pyplot as plt
import numpy as np
probs = np.arange(0,1,0.01)
print(probs)
odds = [p/(1-p) for p in probs]
plt.plot(probs,odds)
plt.xlabel('p')
plt.ylabel('p/(1-p)')
plt.show()
```

ÏãúÍ∑∏Î™®Ïù¥Îìú Ìï®ÏàòÍ∞Ä ÎßåÎì§Ïñ¥ÏßÄÎäî Í≥ºÏ†ï
Î°úÏßìÌï®Ïàò(logit function) (log : Î∞ëÏù¥ eÏù∏ ÏûêÏó∞Î°úÍ∑∏)

  *  pÍ∞Ä 0.5ÏùºÎïå 0Ïù¥ ÎêòÍ≥† pÍ∞Ä 0Í≥º 1ÏùºÎïå Í∞ÅÍ∞Å ÏùåÏùò Î¨¥ÌïúÎåÄÏôÄ ÏñëÏùò Î¨¥ÌïúÎåÄÍ∞Ä Îê®

```python
#Î°úÏßìÌï®Ïàò Ïã§Ïäµ
import matplotlib.pyplot as plt
import numpy as np
probs = np.arange(0.001,0.999,0.001)
print(probs[:100])
print(probs[-100:])
logit = [np.log(p/(1-p)) for p in probs]
plt.plot(probs,logit)
plt.xlabel('p')
plt.ylabel('log(p/(1-p))')
plt.show()
```

```python
#Training dataÏôÄ test dataÎ°ú ÎÇòÎàÑÍ∏∞
#1. train_test_split( ) Ìï®ÏàòÎ•º Ïù¥Ïö©ÌïòÏó¨ Îç∞Ïù¥ÌÑ∞ Î∂ÑÌï†ÌïòÍ∏∞
#: Í∏∞Î≥∏Ï†ÅÏúºÎ°ú training 75%, test 25%Î°ú ÎÇòÎàÑÏñ¥ Ï§å
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import numpy as np
cancer = load_breast_cancer()
x = cancer.data
y = cancer.target
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.2, random_state=42)
# 2. Í≤∞Í≥º ÌôïÏù∏ÌïòÍ∏∞
print(x_train.shape, x_test.shape)

# 3. unique( ) Ìï®ÏàòÎ°ú training dataÏùò target ÌôïÏù∏
print(np.unique(y_train, return_counts=True))
 # ÏñëÏÑ±(1)Ïù¥ ÏùåÏÑ±(0) ÌÅ¥ÎûòÏä§Î≥¥Îã§ 1.7 Î∞∞ Ï†ïÎèÑ ÎßéÏùå
```

```python
# ÎÑòÌååÏù¥Î•º Ïù¥Ïö©Ìïú Ïó∞ÏÇ∞
import numpy as np
a = np.array([1,2,3])
b = np.array([3,4,5])
print(a)
print(b)
print(a+b)
print(a*b)
print(np.sum(a*b))
```

```python
class LogisticNeuron:
  def __init__(self): # Í∞ÄÏ§ëÏπòÏôÄ Ï†àÌé∏ÏùÄ ÎØ∏Î¶¨ Ï¥àÍ∏∞Ìôî ÌïòÏßÄ ÏïäÏùå
    self.w = None
    self.b = None
  def forpass(self,x):
    z = np.sum(x*self.w) + self.b # ÏßÅÏÑ† Î∞©Ï†ïÏãùÏùÑ Í≥ÑÏÇ∞ ,xÏôÄ wÎäî 1Ï∞®Ïõê numpyÎ∞∞Ïó¥, Î∞∞Ïó¥Ïùò ÏöîÏÜåÎÅºÎ¶¨ ÏûêÎèôÍ≥ÑÏÇ∞
    return z
  def backprop(self,err):
    w_grad = x *err #Í∞ÄÏ§ëÏπòÏóê ÎåÄÌïú Í∑∏ÎûòÎîîÏñ∏Ìä∏ Í≥ÑÏÇ∞
    b_grad = 1*err # Ï†àÌé∏Ïóê ÎåÄÌïú Í∑∏ÎûòÎîîÏñ∏Ìä∏ Í≥ÑÏÇ∞
    return w_grad,b_grad
```

Training ÌïòÎäî Î©îÏÑúÎìú Íµ¨ÌòÑ

activaiton() Î©îÏÑúÎìú Íµ¨ÌòÑ
  * ÏãúÍ∑∏Î™®Ïù¥Îìú Ìï®Ïàò ÏÇ¨Ïö©
  * np.exp() : ÏûêÏó∞ÏÉÅÏàòÏùò ÏßÄÏàòÌï®Ïàò

def activation(self, z):
a = 1 / (1 + np.exp(-z)) # ÏãúÍ∑∏Î™®Ïù¥Îìú Í≥ÑÏÇ∞

return a

```python
def fit(self, x, y, epochs=100):
  self.w = np.ones(x.shape[1]) # Í∞ÄÏ§ëÏπòÎ•º 1Î°ú Ï¥àÍ∏∞Ìôî, np.ones( ) Ìï®ÏàòÎäî ÏßÄÏ†ïÌïú ÌÅ¨Í∏∞Ïùò Î∞∞Ïó¥ÏùÑ ÎßåÎì§Í≥† Î™®Îëê 1Î°ú Ï±ÑÏõÄ
  self.b = 0 # Ï†àÌé∏ÏùÑ 0ÏúºÎ°ú Ï¥àÍ∏∞Ìôî
  for i in range(epochs): # epochsÎßåÌÅº Î∞òÎ≥µ
    for x_i, y_i in zip(x, y): # Î™®Îì† ÏÉòÌîåÏóê ÎåÄÌï¥ Î∞òÎ≥µ
      z = self.forpass(x_i) # Ï†ïÎ∞©Ìñ• Í≥ÑÏÇ∞
      a = self.activation(z) # ÌôúÏÑ±Ìôî Ìï®Ïàò Ï†ÅÏö©
      err = -(y_i - a) # Ïò§Ï∞® Í≥ÑÏÇ∞
      w_grad, b_grad = self.backprop(x_i, err) # Ïó≠Î∞©Ìñ• Í≥ÑÏÇ∞
      self.w -= w_grad # Í∞ÄÏ§ëÏπò ÏóÖÎç∞Ïù¥Ìä∏ self.w = self.w - w_grad
      self.b -= b_grad # Ï†àÌé∏ ÏóÖÎç∞Ïù¥Ìä∏ self.b = self.b - b_grad
```

predict( ) Î©îÏÑúÎìú Íµ¨ÌòÑ
* ÏÉàÎ°úÏö¥ ÏÉòÌîåÏóê ÎåÄÌïú ÏòàÏ∏°Í∞í Í≥ÑÏÇ∞
* a>0.5 Ïù¥Î©¥ True (1)Î•º Î∞òÌôòÌïòÍ≥† ÏïÑÎãàÎ©¥ False (0)Î•º Î∞òÌôòÌï®

```python
def predict(self, x):
  z = [self.forpass(x_i) for x_i in x] # Ï†ïÎ∞©Ìñ• Í≥ÑÏÇ∞
  a = self.activation(np.array(z)) # ÌôúÏÑ±Ìôî Ìï®Ïàò Ï†ÅÏö©
  return a > 0.5 # Í≥ÑÎã® Ìï®Ïàò Ï†ÅÏö©
```

```python
#Íµ¨ÌòÑÎêú Ï†ÑÏ≤¥ python ÏΩîÎìú
import numpy as np

class LogisticNeuron:

    def __init__(self):
        self.w = None
        self.b = None

    def forpass(self, x):
        z = np.sum(x * self.w) + self.b  # ÏßÅÏÑ† Î∞©Ï†ïÏãùÏùÑ Í≥ÑÏÇ∞Ìï©ÎãàÎã§.
        return z

    def backprop(self, x, err):
        w_grad = x * err  # Í∞ÄÏ§ëÏπòÏóê ÎåÄÌïú Í∑∏Î†àÎîîÏñ∏Ìä∏Î•º Í≥ÑÏÇ∞Ìï©ÎãàÎã§.
        b_grad = 1 * err  # Ï†àÌé∏Ïóê ÎåÄÌïú Í∑∏Î†àÎîîÏñ∏Ìä∏Î•º Í≥ÑÏÇ∞Ìï©ÎãàÎã§.
        return w_grad, b_grad

    def activation(self, z):
        a = 1 / (1 + np.exp(-z))  # ÏãúÍ∑∏Î™®Ïù¥Îìú Í≥ÑÏÇ∞
        return a

    def fit(self, x, y, epochs=100):
        self.w = np.ones(x.shape[1])  # Í∞ÄÏ§ëÏπòÎ•º Ï¥àÍ∏∞ÌôîÌï©ÎãàÎã§.
        self.b = 0  # Ï†àÌé∏ÏùÑ Ï¥àÍ∏∞ÌôîÌï©ÎãàÎã§.
        for i in range(epochs):  # epochsÎßåÌÅº Î∞òÎ≥µÌï©ÎãàÎã§.
            for x_i, y_i in zip(x, y):  # Î™®Îì† ÏÉòÌîåÏóê ÎåÄÌï¥ Î∞òÎ≥µÌï©ÎãàÎã§.
                z = self.forpass(x_i)  # Ï†ïÎ∞©Ìñ• Í≥ÑÏÇ∞
                a = self.activation(z)  # ÌôúÏÑ±Ìôî Ìï®Ïàò Ï†ÅÏö©
                err = -(y_i - a)  # Ïò§Ï∞® Í≥ÑÏÇ∞
                w_grad, b_grad = self.backprop(x_i, err)  # Ïó≠Î∞©Ìñ• Í≥ÑÏÇ∞
                self.w -= w_grad  # Í∞ÄÏ§ëÏπò ÏóÖÎç∞Ïù¥Ìä∏
                self.b -= b_grad  # Ï†àÌé∏ ÏóÖÎç∞Ïù¥Ìä∏

    def predict(self, x):
        z = [self.forpass(x_i) for x_i in x]  # Ï†ïÎ∞©Ìñ• Í≥ÑÏÇ∞
        a = self.activation(np.array(z))  # ÌôúÏÑ±Ìôî Ìï®Ïàò Ï†ÅÏö©
        return a > 0.5
```

```python
#Î°úÏßÄÏä§Ìã± ÌöåÍ∑ÄÎ™®Ìòï training
#1. Î™®Ìòï training
neuron = LogisticNeuron()
neuron.fit(x_train, y_train)

#2. Íµ¨Ï∂ïÎêú Î™®Ìòï test
print(np.mean(neuron.predict(x_test) == y_test))
```
