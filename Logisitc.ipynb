# 4-1 초기 AI 알고리즘과 로지스틱 회귀


퍼셉트론 (Perceptron)
*   1957년 코넬 항공 연구소(Cornell Aeronautical Lab)의 로젠블라트(F. Rosenblatt)이 제안
*   이진 분류 문제에서 최적의 가중치를 학습하는 알고리즘
*   True / False, 1 / 0
*   퍼셉트론은 사이킷런 패키지에서 제공되는 Perceptron 클래스를 이용 가능

Perceptron 구조
*   마지막 단계에서 샘플의 이진분류를 위한 계단 함수(step function) 사용
*   계단 함수
    𝑦 =  1 (z ≥ 0)
    y = −1 (𝑧 < 0)

*   여러 개의 특성(변수)을 사용하여 수식의 표현 범위를 확장

아달린 (Adaline)
*   1960년 Stanford 대학의 위드로우(B. Widrow)와 호프(T. Hoff)에 의해 연구
*   Perceptron을 개선한 적응형 선형 뉴런(Adaptive Linear Neuron) → Adaline
*   역방향 계산이 계단 함수 출력 이후에 일어나지 않고 선형 함수 출력 이후에 진행됨
*   선형함수의 결과(z)를 학습에 사용
*   계단함수의 결과(𝑦_hat)는 예측에만 활용

로지스틱 휘귀(Logisitc regresesion)
*   Adaline에서 발전된 이진 분류(binary classification) 알고리즘
*   활성화 함수(activation function)을 이용하여 z를 임계 함수에 보내기 전에 활성화 함수를 통과한 a를 계싼

*   임계 함수(threshold function)
  *   Perception의 계단 함수와 비슷한 역할 수행

활성화 함수
*   비선형(non linear)함수 사용

만약 활성화 함수가 선형 함수이면 활성화 함수의 사요이 무의미
*   선형 함수 a와 동일한 구조의 결과(식)이 나오기 때문에 활성화 함수의 의미가 없어짐
*   따라서 활성화 함수는 비선형 함수를 사용해야 함

```python
# 넘파이를 이용한 연산
import numpy as np
a = np.array([1,2,3])
b = np.array([3,4,5])
print(a)
print(b)
print(a+b)
print(a*b)
print(np.sum(a*b))
```

```python
class LogisticNeuron:
  def __init__(self): # 가중치와 절편은 미리 초기화 하지 않음
    self.w = None
    self.b = None
  def forpass(self,x):
    z = np.sum(x*self.w) + self.b # 직선 방정식을 계산 ,x와 w는 1차원 numpy배열, 배열의 요소끼리 자동계산
    return z
  def backprop(self,err):
    w_grad = x *err #가중치에 대한 그래디언트 계산
    b_grad = 1*err # 절편에 대한 그래디언트 계산
    return w_grad,b_grad
```

Training 하는 메서드 구현

activaiton() 메서드 구현
  * 시그모이드 함수 사용
  * np.exp() : 자연상수의 지수함수

def activation(self, z):
a = 1 / (1 + np.exp(-z)) # 시그모이드 계산

return a

```python
def fit(self, x, y, epochs=100):
  self.w = np.ones(x.shape[1]) # 가중치를 1로 초기화, np.ones( ) 함수는 지정한 크기의 배열을 만들고 모두 1로 채움
  self.b = 0 # 절편을 0으로 초기화
  for i in range(epochs): # epochs만큼 반복
    for x_i, y_i in zip(x, y): # 모든 샘플에 대해 반복
      z = self.forpass(x_i) # 정방향 계산
      a = self.activation(z) # 활성화 함수 적용
      err = -(y_i - a) # 오차 계산
      w_grad, b_grad = self.backprop(x_i, err) # 역방향 계산
      self.w -= w_grad # 가중치 업데이트 self.w = self.w - w_grad
      self.b -= b_grad # 절편 업데이트 self.b = self.b - b_grad
```

  * np.zeros()
  * np.full() : 0과 1이 아닌 임의의 값으로 배열을 생성

예측하는 메서드 구현

predict( ) 메서드 구현
* 새로운 샘플에 대한 예측값 계산
* a>0.5 이면 True (1)를 반환하고 아니면 False (0)를 반환함

```python
def predict(self, x):
  z = [self.forpass(x_i) for x_i in x] # 정방향 계산
  a = self.activation(np.array(z)) # 활성화 함수 적용
  return a > 0.5 # 계단 함수 적용
```

```python
#구현된 전체 python 코드
import numpy as np

class LogisticNeuron:

    def __init__(self):
        self.w = None
        self.b = None

    def forpass(self, x):
        z = np.sum(x * self.w) + self.b  # 직선 방정식을 계산합니다.
        return z

    def backprop(self, x, err):
        w_grad = x * err  # 가중치에 대한 그레디언트를 계산합니다.
        b_grad = 1 * err  # 절편에 대한 그레디언트를 계산합니다.
        return w_grad, b_grad

    def activation(self, z):
        a = 1 / (1 + np.exp(-z))  # 시그모이드 계산
        return a

    def fit(self, x, y, epochs=100):
        self.w = np.ones(x.shape[1])  # 가중치를 초기화합니다.
        self.b = 0  # 절편을 초기화합니다.
        for i in range(epochs):  # epochs만큼 반복합니다.
            for x_i, y_i in zip(x, y):  # 모든 샘플에 대해 반복합니다.
                z = self.forpass(x_i)  # 정방향 계산
                a = self.activation(z)  # 활성화 함수 적용
                err = -(y_i - a)  # 오차 계산
                w_grad, b_grad = self.backprop(x_i, err)  # 역방향 계산
                self.w -= w_grad  # 가중치 업데이트
                self.b -= b_grad  # 절편 업데이트

    def predict(self, x):
        z = [self.forpass(x_i) for x_i in x]  # 정방향 계산
        a = self.activation(np.array(z))  # 활성화 함수 적용
        return a > 0.5
```

```python
#로지스틱 회귀모형 training
#1. 모형 training
neuron = LogisticNeuron()
neuron.fit(x_train, y_train)

#2. 구축된 모형 test
print(np.mean(neuron.predict(x_test) == y_test))
```
