## 1.가설 검정의 절차 가설의 종류

* 가설검정 : 모집단에 대해 가설 설정 후, 표본관찰을 통해 그 가설의 채택 여부를 결정하는 통계적 추론 방법

* 가설검정의 절차
image.png


가설의 종류
- 귀무가설
  * 가설검정의 대상이 되는 가설, 연구자가 부정하고자 하는 가설
  * 알고 있는 것과 같음, 변화 없음, 영향력 없음 , 연관성 없음, 효과 없음에 대하 가설

- 대립가설
  * 연구자가 연구를 통해 입증/증명되기를 기대하는 예상이나 주장
  * 귀무가설이 기각되면 채택되는 사실
  * 알고 있는 것과 다름, 변화 있음, 영향력 있음, 연관성 있음, 효과 있음에 대한 가설



## 2.가설검정의 종류

* 가설검정의 종류 : 귀무가설을 기각하는 영역의 위치에 따라 세 가지로 나눔

- 양측 검증
  * 귀무가설을 기각하는 영역이 양쪽에 있는 검정

- 단측 검증
  * 좌측 검증
    * 귀무가설을 기각하는 영역이 왼쪽에 있는 검정
  * 우측 검증
    * 귀무가설을 기각하는 영역이 오른쪽에 있는 검정

## 3.귀무가설,대립가설

귀무가설, 대립가설 설정
  * 귀무가설은 "같다,이상,이하"의 세가지 유형이 있음
  * 귀무가설의 반대인 대립가설은 '다르다,작다(미만), 크다(초과)"의 세가지 유형이 있음
  * 귀무가설 및 대립가설은 모집단의 모수에 대한 추론으로 ,σ,σ^2를 가설에 사용함


핸드폰 교체주기는 2년이라고 알려져 있는데, 어떤 점눈가는 2년이 아니라고 주장한다, 이에 핸드폰 평군 교체주기가 실제로 어떻게 되는지 검증하력 한다. 가설을 설정해보자

* H0 : μ=2 (평균 교체주기가 2년이다)  
* H1 : μ:=/2 (평균 교체주기가 2년이 아니다)

image.png


A 학교 성적의 분산이 7이상으로 알려져 있다. 그런데 기말고사 성적에 따르면 분산이 7보다 작을 수 있다는 의견이 나왔다. 실제 분산에 대해 검정하려 한다. 가설을 설정해보자

* H0 : σ^2 ≥ 7 (분산이 7보다 크거나 같다)
* H1 : σ^2 <  7 (분산이 7보다 작다)


image.png

X 매장의 제품 분실율은 5%이하로 알려져 있다. 그러나 재고 조사 결과 분실욜이 5%보다 더 클 수 있다는 의견이 나왔다. 이에 실제 분실율에 대해 검정하려 한다. 가설을 설정해보자

* H0 : P ≤ : 0.05 ( 분실율은 5%보다 작거나 같다)
* H1 : P > : 0.05 ( 분실율은 5%보다 크다)

## 4.제1종,2종 오류

* 제 1종 오류(a error)
  * 귀무가설이 참인데 기각되는 오류, a = P (제1종 과오를 범함) = P(H0 기각 | H0 참)

* 제 2종 오류(B error)
  * 귀무가설이 거짓인데 채택하는 오류,B=P(제 2종 과오를 범함) = P (H0 채택| H0 거짓)

* 신뢰수준(1-a)
  * 1종 오류를 범하지 않을 확률
* 검정력(1-B)
  * 2종 오류를 범하지 않을 확률

image.png

* 두가지 오류가 작을수록 바람직함
* a 와 B (1-a)와 (1-B)는 각각 Trade off 관계
* 표본의 크기를 증가 시키면 표본분산이 작이지게 되며, a,B 발생 확률이 작아지지만 조사비용 증가
* 제1종 오류를 범할 확률의 최대 허용치를 특정값(유의수준)으로 지정해 놓고 제 2종 오류의 확률을 가장 작게 해주는 검정 방법을 사용


## 5.신뢰수준,기각역,유의수준

* 신뢰수준(1-a)
  * 제 1종 오류를 범하지 않을 확률, 연구활동(99%),일반적(95%),단순설문조사(90%) 사용

* 기각역
  * 귀무가설을 기각하고 대립(=연구) 가설을 채택하는 되는 영역

* 유의수준(a)
  * 귀무가설이 참인데도 기각시키는 확률(제 1종 오류 발생 확률)의 최대 허용 한계
  * 유의수준 0.05(5%) : 100번 실험에서 제1종 오류를 범하는 최대 허용 한계가 5번
  * 유의수준 = 1- 신뢰수준

## 6. 검정통계량

* 검정통계량 : 귀무가설을 채택할지 기각할지를 정하는 판단 기준이 되는 통계량(그래프 x축 값)

* 확률분포에 따라 t,Z,F,X^2 통계량이 있음

* 분포표를 사용하여 자유도를 고려한 검정통께량,확률값(p-value)을 구할 수 있음

image.png

## 7. 유의확률 (p-value),귀무가설 기각/채택


* 유의확률(p-value) : Probability  Value, 0≤p-value≤1
* 자유도를 고려한 검정통계량에 관한 확률
* 귀무가설의 신뢰구간을 벗어나는 확률(=극단적인 표본 값이 나올 확률), 판정이 잘못되었을 확률
* 제1종 오류를 범할 확률, 귀무가설을 지지하는 정도
* p-value가 작을 수록 그 정도가 약하다고 보며, p-value < a 일때, 귀무가설을 기각, 대립가설을 채택함
* p-value가 0.05(5%)
  * 귀무가설을 기각했을 때 기각 결정이 잘못될 확률이 5%임
