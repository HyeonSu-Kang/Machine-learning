mtcars 데이터셋(mtcars.csv)의 qsec 컬럼을 최소 최대 척도(Min-Max Scale)로 변환후 0.5보다 큰값을 가지는 레코드 수를 구하시오

```python
#[1] MinMaxSclaer 사용
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/main/mtcars.csv')
df.head(2)
from sklearn.preprocessing import MinMaxScaler
df['qsec'] = MinMaxScaler().fit_transform(df[['qsec']])
count = df['qsec'] > 0.5
print(count.sum())

#[2] StandardScaler 사용
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/main/mtcars.csv')
df.head(2)
from sklearn.preprocessing import StandardScaler
temp = StandardScaler().fit_transform(df[['qsec']])
# 표준정규분포(평균:0,표준편차:1)
print(temp.mean(),temp.std(ddof=0))
```

```python
#[3] min_max_scaler 직접구현하는 방법
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/main/mtcars.csv')
df.head(2)
def min_max_scaler(a):
  return(a-a.min())/(a.max()-a.min())
temp = min_max_scaler(df['qsec'])
print(temp.min(),temp.max())

result1 = len(temp[temp>0.5])
result2 = temp[temp>0.5].shape[0]
result3 = (temp>0.5).sum()
print(result1,result2,result3)

# StandardScaler 함수 생성
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/main/mtcars.csv')

def StandardScaler(a):
  return(a - a.mean())/a.std(ddof=0)

temp = StandardScaler(df['qsec'])
print(temp.mean(),temp.std(ddof=0))
```

```python
#[1] 보스턴 데이터 범죄율 컬럼('CRIM')의 top10 구하기
A = df['CRIM'].sort_values(ascending=False).head(10)
#[2] 10번째 범죄율 값으로 1~10위의 범죄율 값을 변경하기
df.loc[A.index,'CRIM'] = A.iloc[-1]
#[3] (전체 데이터에서)'AGE' 변수의 값이 80 이상인 것에 대한 범죄율 평균 구하기
result1 = df.loc[df['AGE'] >= 80,'CRIM'].mean()
result2 = df.loc[df.AGE >=80,'CRIM'].mean()
print(result1,result2)
```

#3.하우징 데이터

###3.1하우징 데이터

* 주어진 데이터의 첫번째 행 부터 순서대로 80%까지의 데이터를 추출
* total_bedrooms 변수의 결측값(NA)을 'total_bedrooms 변수의 중앙값으로 대체하고,
* 대체 전의 total_bedrooms 변수 표준편차과 대체 후의 'total_bedrooms'변수 표준편차 값을 산출하여
* 대체 전 표준편차 값,대체 후 표준편차 값을 구해 출력한다.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/main/housing.csv')
print(df.shape)
df.head(2)
```

```python
#[1] 계산 사용
#!) 첫번째 행 부터 순서대로 80% 까지의 데이터를 추출
p = int(df.shape[0]*0.8)
#p = round(df.shape[0]*0.8)
A = df.loc[:p]

#2) total_bedrooms 변수의 결측값(NA)를 total_bedrooms 변수의 중앙값으로 대체
A_before = A['total_bedrooms']
A_after = A['total_bedrooms'].fillna(A['total_bedrooms'].median())

#3) 대체 전 표본편차 값 - 대채 후 표분편차 값
result1 = A_before.std(ddof=0) - A_after.std(ddof=0)
result2 = A_before.std(ddof=1) - A_after.std(ddof=1)
print(f'{result1:.4f},{result2:.4f}')
print(round(result1),round(result2))
```

###3.2 하우징 데이터

* 하우징 데이터에서 'latitude' 칼럼의 이상치를 찾아 이상치들의 합을 산출하시오
* 이상치 기준 : 평군 - (표준편차*1.5), 평균 + (표준편차*1.5)



```python
#[0] 데이터준비
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/main/housing.csv')
df.head(2)
```

```python
# 이상치 합 구하기
temp = df['latitude']
t_mean = temp.mean()
t_std = temp.std(ddof=0)

lower = t_mean-1.5*t_std
upper = t_mean+1.5*t_std
a = temp[temp>upper].sum()
b = temp[temp<lower].sum()
print(a+b)
```

```python
# 이상치에 해당하는 데이터 개수
df = pd.read_csv('https://raw.githubusercontent.com/HyeonSu-Kang/data/main/housing.csv')
temp = df['latitude']
mean,std = temp.mean(),temp.std(ddof=0)
lower = mean - 1.5*std
upper = mean + 1.5*std
a = temp[temp<lower].shape[0]
b = temp[temp>upper].shape[0]
result = a+b
print(result)

cond = (temp < lower) | (temp > upper)
result2 = temp[cond].shape[0]
print(result2)

# 이상치에 해당하는 데이터 검색
cond = (temp<lower) | (temp>upper)
df[cond]
```
