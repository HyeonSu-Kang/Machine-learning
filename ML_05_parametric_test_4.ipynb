# T-test


##1.T-test

*  스튜던트 t-테스트(student's t-test)라고도 하며, 검정 통계량이 귀무가설 하에서 t-분포를 따른 통계적 가설 검정

* 표본을 사용한 모평균 검정 및 데이터 세트(집단)의 모평균이 서로 유의하게 다른지 여부를 판별 할 수 있음

* 검정통께량이 정규 분포를 따르며 모집단의 분산,표준편차를 알지 못할 때 표본으로 부터 추정된 분산/표준편차를 사용해 검정함

* t-test를 실시하기 위해서는 정규성 및 등분산성의 조건이 만족되어야 함

##3.t-test를 위한 함수


* 검정방법
  * One sample Test,1표본 평균 검정
    * test_1samp(a, popmean, alternative-'two-sided') statistic, p-value
    * https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest Isamp.html
  * Paired T test,쌍체(대응) 표본 t-검정
    * ttest_rel(a, b, alternative-'two-sided') statistic, p-value
    * https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_rel.html
  
  * Two sample Test,2표본 평균 검정

    * ttest ind(a, b,alternative='two-sided", equal_var-True) → statistic, p-value equal_var=False 인 경우 Welch's t-test를 수행함
    * https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest ind.html

* alternative(대립가설)
  * two-sided
    * 두집단의 평균이 다르다
  * less
    * 첫 번째 집단의 평균이 두번째 집단의 평균보다 작다
  * greater
    * 첫 번째 집단의 평균이 두 번째 집단의 평균보다 크다

* alternative less - (효과 있음에 대해 판단할 때)
  * ex) ttest_rel(df['after'].df['before'],alternative='less')
