### 1-04.서식있는 문자열

```python
# [1] 1000 값을 갖는 정수 객체 a를 사용하여 'Price is 1000'을 출력한다
# % 연산자, format 메서드, f-string 사용 => 문자열 만드는 방법
a=1000
print('Price is',a)
print('Price is %d' %(a))
print('price is {}'.format(a))
print(f'Price is {a}')
```

### 1-05.연산자

```python
# [1] 거듭제곱, 산술 연산자
a, b = 10, 3

# [1-1] a의 b승을 연산자와 함수를 사용하여 구해본다
print(a**b,pow(a,b))
# [1-2] a와 b의 사칙연산(+, -, *, /)를 실행해 본다
#       (나눗셈의 결과는 반올림하여 소수점아래 2개까지 표시)
print(a+b,a*b,a-b,f'{a/b:.2f}')

# [1-3] a와 b를 나누어 그 몫, 나머지를 연산자와 함수를 사용하여 구해본다
print(a//b,a%b,divmod(a,b))
```

```python
# [2] (  )를 사용한 우선 순위 변경
#     b와 c를 더한 결과를 a와 곱한 결과 구하기
a, b, c = 10, 3, 1
d = a * b + c
e = a * (b + c)
print(d)
print(e)
```

```python
# [3] 비교 연산자 (==, !=, >, <, >=, <=) : 결과는 True/False
a, b = 10, 5

# [3-1] a와 b가 같다, 다르다의 결과 출력
print(a==b)

# [3-2] a가 b보다 크다(초과), 작다(미만)의 결과 출력
print(a>b,a<b)

# [3-3] a가 b보다 크거나 같다(이상), 작거나 같다(이하)의 결과 출력
print(a>=b,a<=b)

# [3-4] a가 5보다 크고, 15보다 작다
print(15>a>5)
```

```python
# [4] a, b의 값 비교(==), 주소 비교(is), 포함된 것인지 확인 (in) : 결과는 True/False
a = 300
b = 300
print(id(a),id(b) )# 둘은 다른 객체이다
print(a==b, a is b)
print(a in [10,20,30], a in [100,200,300])
```

```python
# [5] 논리 연산자 (and, or, not)
a = 10

# [5-1] a가 정수이고, 10보다 크다
print(type(a) is int and a > 10)

# [5-2] a가 정수이거나, 10보다 크다
print((type(a) is int) or a>10)

# [5-3] a가 0이다
print(a==0,not a)
```

```python
# [6] 연산자 활용 -1
# 다음 숫자에 대해 각각 1, 2, 3이 출력되도록 {} 내부에 식을 작성하여 보도록 한다
c = 123
print(f'백의 자리 = {c//100}, 십의 자리 = {c//10%10}, 일의 자리 ={c%10}')
```

```python
# [7] 연산자 활용 -2
# a의 값을 1 증가시켜 출력해 보자
a = 10

print(a+1)

a += 1
print(a)
```

```python
# [8] Numeric type 끼리는 type이 달라도 연산 가능
# 표현 범위가 넓은 type의 결과를 생성함  float > int > bool
# int + float = float, int + bool = int
a = 10 + 3.14
b = 1 + True

# a, b의 결과와 객체 type은 무엇일까?
print(a, type(a))
print(b, type(b))
```

```python
# [9] Container는 연산자 사용시 같은 type 끼리만 연산 가능
# 필요시 container 형변환 list(b)
a = [1, 2, 3]
b = (4, 5)

c = a + list(b)
d = a*3
print(c,d,sep='\n')
```


### 1-09.함수(function)
- 함수 :  입력 데이터를 가지고 처리하여 결과를 반환하는 명령의 묶음
- 함수의 구성 : parameter, operation, return
- built-in 함수, 외부 함수 등 다양하게 준비된 함수가 있음
- 사용자가 만들어 사용할 수 있음
- 함수 vs 메서드(method)
   - 함수는 module(=파일)에서 가지고 있는 기능을 의미한다
   - 메서드는 class 내부에 작성된 함수이다
   - 함수와 메서드에 접근하기 위해서는 '.' 연산자를 사용하여 접근하도록 한다

```python
def func(a, b):
    c = a + b
    d = a - b
    return c, d

func(10, 30)
```
